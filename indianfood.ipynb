```python
from nbformat import v4 as nbf

# Create a new notebook
nb = nbf.new_notebook()

# List of cells to include
cells = []

# Install cell
cells.append(nbf.new_code_cell("""\
!pip install gradio pandas
"""))

# Dataset creation cell
cells.append(nbf.new_code_cell("""\
import pandas as pd

# Sample data
data = {
    "name": ["Aloo Paratha", "Dosa", "Rasgulla", "Masoor Dal", "Idli"],
    "ingredients": [
        "Potatoes, Flour, Ghee, Spices",
        "Rice, Urad Dal, Oil",
        "Chhena, Sugar, Water",
        "Red Lentils, Onion, Tomato, Spices",
        "Rice, Urad Dal, Salt"
    ],
    "prep_time": [20, 480, 15, 10, 480],
    "cook_time": [15, 10, 30, 20, 10],
    "flavor_profile": ["spicy", "savory", "sweet", "spicy", "savory"],
    "course": ["main course", "main course", "snack", "main course", "snack"],
    "region": ["North India", "South India", "East India", "North India", "South India"]
}

df = pd.DataFrame(data)
df.to_csv("IndianFoodDataset.csv", index=False)
df.head()
"""))

# Gradio chatbot app code cell
cells.append(nbf.new_code_cell("""\
import gradio as gr

# Load the dataset
df = pd.read_csv("IndianFoodDataset.csv")

# Define chatbot logic
def generate_recipe(user_input):
    filtered = df.copy()
    query = user_input.lower()

    if "north" in query:
        filtered = filtered[filtered['region'].str.lower() == "north india"]
    elif "south" in query:
        filtered = filtered[filtered['region'].str.lower() == "south india"]
    elif "east" in query:
        filtered = filtered[filtered['region'].str.lower() == "east india"]

    if "snack" in query:
        filtered = filtered[filtered['course'].str.lower() == "snack"]
    elif "main" in query:
        filtered = filtered[filtered['course'].str.lower() == "main course"]

    if "sweet" in query:
        filtered = filtered[filtered['flavor_profile'].str.lower() == "sweet"]
    elif "spicy" in query:
        filtered = filtered[filtered['flavor_profile'].str.lower() == "spicy"]
    elif "savory" in query:
        filtered = filtered[filtered['flavor_profile'].str.lower() == "savory"]

    if not filtered.empty:
        recipe = filtered.sample(1).iloc[0]
        return f\"\"\"
üçõ **{recipe['name']}**

**Region**: {recipe['region']}
**Course**: {recipe['course']}
**Flavor**: {recipe['flavor_profile']}

**Ingredients**: {recipe['ingredients']}
**Prep Time**: {recipe['prep_time']} min
**Cook Time**: {recipe['cook_time']} min
\"\"\"
    else:
        return "‚ùå Sorry, no matching recipe found. Try another combination."

# Launch Gradio interface
gr.Interface(
    fn=generate_recipe,
    inputs=gr.Textbox(placeholder="e.g., spicy South Indian dish"),
    outputs="markdown",
    title="üáÆüá≥ Indian Recipe Generator",
    description="Get delicious Indian recipes by typing things like 'spicy South Indian main dish'"
).launch(share=True)
"""))

# Add cells to notebook and write to file
nb['cells'] = cells
notebook_path = "/mnt/data/Indian_Recipe_Chatbot_Gradio_Colab.ipynb"

with open(notebook_path, 'w') as f:
    nbf.write(nb, f)

notebook_path
```
